# goal
Build a console-based shopping experience where the user can:

Browse available products

Add items to their cart

Remove items if needed

See the cart and total bill

Checkout and exit

```python
print("----------WELCOME---------")
print("Have a great day")
print("How can I help you today?")

items = {"shoes": 1200, "bags": 850, "skincare": 2000, "clothes": 2500}
cart = {}

while True:
    print("\n1) View products")
    print("2) Add to cart")
    print("3) View cart")
    print("4) Checkout")
    print("5) Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("\nAvailable Products:")
        for item, price in items.items():
            print(f"{item.title()} = â‚¹{price}")

    elif choice == 2:
        n = input("Enter your item: ").lower()
        if n in items:
            qua = int(input("Enter quantity: "))
            if n in cart:
                cart[n] += qua
            else:
                cart[n] = qua
            print(f" Added {qua} {n}(s) to your cart")
        else:
            print(" No such item found!")

    elif choice == 3:
        if not cart:
            print("Your cart is empty.")
        else:
            print("Your Cart:")
            total = 0
            for item, qty in cart.items():
                price = items[item]
                cost = price * qty
                total += cost
                print(f"{item.title()} x {qty} = â‚¹{cost}")
            print(f"Total Amount: â‚¹{total}")

    elif choice == 4:
        if not cart:
            print("Your cart is empty.")
        else:
            print("\nFinal Bill:")
            total = 0
            for item, qty in cart.items():
                price = items[item]
                cost = price * qty
                total += cost
                print(f"{item.title()} x {qty} = â‚¹{cost}")
            print(f"\n Grand Total: â‚¹{total}")
            print("Thank you for shopping at Varshi's store!")
            cart.clear()

    elif choice == 5:
        print("ðŸ‘‹ Thanks for visiting! Come again.")
        break

    else:
        print("Invalid,Please try again.")
```
