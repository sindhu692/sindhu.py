```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"F:\sindhu\python\patient.csv")
print("Original report:")
print(df)
#avg days admitted
for i in df["Admitted_Days"]:
    print(i,end=" ")
avg_days=df["Admitted_Days"].mean()
print("avg days admitted:",round(avg_days,2))
#max bill
max_patient=df.loc[df["Bill"].idxmax()]
print(f"\n Highest bill paid by {max_patient['Name']}-{max_patient['Bill']}")
#list patients above 60
print(f"\n Age above 60:\n",df[df['Age']>60][['Name','Disease','Bill']])
#Disease summary:
print(f"\n Diesease Summary:",df["Disease"].value_counts())

#plots
#subplot of patient and bill
fig = plt.figure(figsize=(14,6))
plt.subplot(2,1,1)
plt.bar(df["Name"],df["Bill"],color='g')
plt.title("Patients Bill Summary")
plt.xlabel("Patients")
plt.ylabel("Bill Amount ()")
plt.grid(axis='y',linestyle='--',alpha=0.6)

#subplot
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd plot
disease_counts = df["Disease"].value_counts()
plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%', startangle=90)
plt.title(" Disease Distribution")
plt.axis('equal')

plt.tight_layout()
plt.show()
```
